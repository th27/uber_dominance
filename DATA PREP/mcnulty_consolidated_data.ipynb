```python
import pandas as pd
import numpy as np

import pickle
```

### Import consols

```python
df_ct = pd.read_csv('geoid_census_tract.csv')
df_ct = df_ct[['GEOID','TRACT']]
df_ct.head()
```

```python
with open('target_for_model.pkl','rb') as picklefile:
    df_1 = pickle.load(picklefile)
    
with open('airbnb_for_model.pkl','rb') as picklefile:
    df_2 = pickle.load(picklefile)
    
with open('demographic_for_model.pkl','rb') as picklefile:
    df_3 = pickle.load(picklefile)
    
with open('noisecomplaint_for_model.pkl','rb') as picklefile:
    df_4 = pickle.load(picklefile)

with open('yelpbars_for_model.pkl','rb') as picklefile:
    df_5 = pickle.load(picklefile)
    
with open('yellow_growth_for_model.pkl','rb') as picklefile:
    df_6 = pickle.load(picklefile)
```

```python
df_1.head()
```

```python
df1 = pd.merge(df_ct, df_1, left_on='GEOID', right_on='GEOID', how='left')
```

```python
df1.head()
```

```python
df_2.head()
```

```python
df2 = pd.merge(df1, df_2, left_on='GEOID', right_on='airbnb_census_tract', how='left')
df2.drop(['airbnb_census_tract','TRACT'],axis=1,inplace=True)
```

```python
df2.head()
```

```python
df_3.head()
```

```python
df3 = pd.merge(df2, df_3, left_on='GEOID', right_on='demo_census_tract', how='left')
df3.drop(['demo_census_tract'],axis=1,inplace=True)
```

```python
df3.head()
```

```python
df_4.head()
```

```python
df4 = pd.merge(df3, df_4, left_on='GEOID', right_on='census_tract', how='left')
df4.drop(['census_tract'],axis=1,inplace=True)
```

```python
df4.head()
```

```python
df_5.head()
```

```python
df5 = pd.merge(df4, df_5, left_on='GEOID', right_on='yelp_bar_census_tract', how='left')
df5.drop(['yelp_bar_census_tract'],axis=1,inplace=True)
```

```python
df5.head()
```

```python
df_6.head()
```

```python
df6 = pd.merge(df5, df_6, left_on='GEOID', right_on='census_tract', how='left')
df6.drop(['census_tract'],axis=1,inplace=True)
```

```python
df6.columns
```

```python
df = df6.copy()
```

```python
len(df.columns)
for i in list(range(len(df.columns))):
    print(df.columns[i], df.ix[:,i].isnull().sum())
```

```python
# get rid of rows where there is no uber data
```

```python
df = df[df['share_uber']>=0]
df.shape
```

```python
# count NaNs in rows and get rid of columns with lots of NaN
```

```python
# convert NaNs to zeros
```

```python
df = df.fillna(0)
```

```python
len(df.columns)
for i in list(range(len(df.columns))):
    print(df.columns[i], df.ix[:,i].isnull().sum())
```

```python
with open('master_data.pkl', 'wb') as picklefile:
    pickle.dump(df,picklefile)
```
